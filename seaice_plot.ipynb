```python
import altair as alt
```

```python
chart = alt.Chart(df[df['year'].isin([1979,1980,2010,2012,2015,2016,2018,2019])]).mark_line().encode(x='yearday',
                                                           y=alt.Y('extent', scale=alt.Scale(zero=False)),
                                                           color='year:N',
                                                           tooltip=['year', 'extent'])
```
